```python
%pip install tabulate -q

import pandas as pd
import numpy as np
import os

# 1. Configuración de archivos y rutas
path = r"/content/"
files = {
    "anton_532": "anton_532.csv",
    "anton_785": "anton_785.csv",
    "kaiser": "kaiser.csv",
    "metrohm": "metrohm.csv",
    "mettler_toledo": "mettler_toledo.csv",
    "tec5": "tec5.csv",
    "timegate": "timegate.csv",
    "tornado": "tornado.csv"
}



# Lista para guardar los resultados
datos_resumen = []

# 2. Iterar sobre cada archivo
for nombre_equipo, nombre_archivo in files.items():
    ruta_completa = os.path.join(path, nombre_archivo)

    try:
        # Leemos el CSV
        df = pd.read_csv(ruta_completa)

        # 3. Separar datos espectrales de los metadatos
        # Las últimas 5 columnas son targets/info, el resto son los Shifts
        columnas_totales = df.columns
        columnas_target = columnas_totales[-5:]
        columnas_shift = columnas_totales[:-5] # Estos son los "Raman Shifts"

        # Convertimos los nombres de las columnas a float para hacer cálculos numéricos
        # (Asumiendo que los headers son cadenas de texto representando números)
        shifts_numericos = columnas_shift.astype(float)

        # 4. Calcular métricas
        n_muestras = df.shape[0]
        min_shift = shifts_numericos.min()
        max_shift = shifts_numericos.max()

        # Resolución: Calculamos el paso promedio entre puntos consecutivos
        pasos = np.diff(shifts_numericos)
        resolucion_media = np.mean(pasos)

        # Guardamos en la lista
        datos_resumen.append({
            "Dispositivo": nombre_equipo,
            "Dimensiones (Filas, Cols)": df.shape,
            "Num. Muestras": n_muestras,
            "Min Shift ($cm^{-1}$)": round(min_shift, 2),
            "Max Shift ($cm^{-1}$)": round(max_shift, 2),
            "Resolución Promedio ($cm^{-1}$)": round(resolucion_media, 4),
            "Num. Puntos Espectrales": len(shifts_numericos)
        })

    except Exception as e:
        print(f"Error procesando {nombre_equipo}: {e}")

# 5. Crear el DataFrame final
tabla_resumen = pd.DataFrame(datos_resumen)

# Mostrar la tabla (fallback a print si to_markdown falla)
try:
    print(tabla_resumen.to_markdown(index=False))
except Exception:
    print(tabla_resumen)
```

```python
import numpy as np
import pandas as pd
import os
import pickle
import matplotlib.pyplot as plt
import seaborn as sns

# Modelos potentes
from sklearn.cross_decomposition import PLSRegression
from xgboost import XGBRegressor

# Herramientas de selección y métricas
from sklearn.model_selection import GridSearchCV, PredefinedSplit
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Ignorar advertencias molestas
import warnings
warnings.filterwarnings('ignore')

print("Todo importado correctamente.")
```

```python
# ---------------------------------------------------------------
# PON AQUÍ LA RUTA DONDE ESTÁN TUS ARCHIVOS CSV
# ---------------------------------------------------------------
# Ejemplo Kaggle: '/kaggle/input/dig-4-bio-raman-transfer-learning-challenge'
# Ejemplo Windows: r'C:\Usuarios\TuUsuario\Documentos\Raman'
path = r"/content/"
DATA_PATH = {
    "anton_532": "anton_532.csv",
    "anton_785": "anton_785.csv",
    "kaiser": "kaiser.csv",
    "metrohm": "metrohm.csv",
    "mettler_toledo": "mettler_toledo.csv",
    "tec5": "tec5.csv",
    "timegate": "timegate.csv",
    "tornado": "tornado.csv"
}

print(f"Ruta configurada: {DATA_PATH}")
```

```python
# Configuración de límites espectrales
dataset_names = ['anton_532', 'anton_785', 'kaiser', 'mettler_toledo', 'metrohm', 'tec5', 'timegate', 'tornado']
lower_bounds = {'anton_532': 200, 'anton_785': 100, 'kaiser': -37, 'mettler_toledo': 300, 'metrohm': 200, 'tec5': 85, 'timegate': 200, 'tornado': 300}
upper_bounds = {'anton_532': 3500, 'anton_785': 2300, 'kaiser': 1942, 'mettler_toledo': 3350, 'metrohm': 3350, 'tec5': 3210, 'timegate': 2000, 'tornado': 3300}

def get_csv_dataset(dataset_name, lower_wn=-1000, upper_wn=10000, dtype=None):
    lower_wn = max(lower_wn, lower_bounds[dataset_name])
    upper_wn = min(upper_wn, upper_bounds[dataset_name])
    dtype = dtype or np.float64

    df = pd.read_csv(os.path.join(path, DATA_PATH[dataset_name]))

    # Selección de columnas espectrales
    wns_cols = np.array([float(one) for one in df.columns[:-5]])
    spectra_selection = np.logical_and(lower_wn <= wns_cols, wns_cols <= upper_wn)

    spectra = df.iloc[:, :-5].iloc[:, spectra_selection].values
    label = df.iloc[:, -5:-1].values
    cv_indices = df.iloc[:, -1].values # Folds originales del dispositivo

    wavenumbers = wns_cols[spectra_selection]

    # Crear folds (aunque usaremos nuestra propia estrategia luego)
    all_indices = np.array(range(len(cv_indices)))
    cv_folds = [(all_indices[cv_indices != fold_idx], all_indices[cv_indices == fold_idx]) for fold_idx in range(len(set(cv_indices)))]

    return spectra.astype(dtype), label.astype(dtype), None, cv_folds, wavenumbers.astype(dtype)

def load_joint_dataset(dataset_names, lower_wn=-1000, upper_wn=10000):
    lower_wn = max(lower_wn, *[lower_bounds[name] for name in dataset_names])
    upper_wn = min(upper_wn, *[upper_bounds[name] for name in dataset_names])
    print(f"Rango Común: {lower_wn} - {upper_wn} cm-1")

    datasets = [get_csv_dataset(name, lower_wn=lower_wn, upper_wn=upper_wn) for name in dataset_names]
    joint_wns = np.arange(lower_wn, upper_wn + 1)

    # Interpolación
    interpolated_data = [
        np.array([np.interp(joint_wns, xp=wns, fp=spec) for spec in spectra])
        for spectra, _, _, _, wns in datasets
    ]

    # Normalización Máxima (Crucial para mezclar dispositivos)
    normed_spectra = np.concatenate([spec / np.max(spec, axis=1, keepdims=True) for spec in interpolated_data], axis=0)
    labels = np.concatenate([d[1] for d in datasets])[:, :3] # Solo las 3 etiquetas

    return normed_spectra, labels

def load_comp_data(filepath, is_train=True):
    if is_train:
        df = pd.read_csv(filepath)
        y = df[['Glucose (g/L)', 'Sodium Acetate (g/L)', 'Magnesium Acetate (g/L)']].dropna().values
        X = df.iloc[:, :-4]
    else:
        df = pd.read_csv(filepath, header=None)
        X = df
        y = None

    # Limpieza
    X.columns = ["sample_id"] + [str(i) for i in range(X.shape[1]-1)]
    X['sample_id'] = X['sample_id'].ffill()

    cols = X.columns[1:]
    for col in cols:
        X[col] = pd.to_numeric(X[col].astype(str).str.replace('[','',regex=False).str.replace(']','',regex=False), errors='coerce')

    # Promediar escaneos y reshape
    X_vals = X.drop('sample_id', axis=1).values.reshape(-1, 2, 2048).mean(axis=1)
    return X_vals, y

def fix_val_test_shape(X):
    # Interpolar al mismo rango que el entrenamiento (300-1942)
    joint_wns = np.arange(300, 1942 + 1)
    orig_wns = np.linspace(65, 3350, 2048)
    mask = (orig_wns >= 300) & (orig_wns <= 1942)

    X_crop = X[:, mask]
    wns_crop = orig_wns[mask]

    # Interpolación final
    X_final = np.array([np.interp(joint_wns, xp=wns_crop, fp=spec) for spec in X_crop])
    return X_final

print("Funciones definidas.")
```

```python
# 1. Cargar Entrenamiento (8 Dispositivos)
print("--- Cargando X_train (8 dispositivos) ---")
X_train, y_train = load_joint_dataset(dataset_names)

# 2. Cargar Validación (Transfer Plate)
print("\n--- Cargando X_val (Transfer Plate) ---")
X_val_raw, y_val = load_comp_data(os.path.join(path, 'transfer_plate.csv'), is_train=True)
X_val = fix_val_test_shape(X_val_raw)
X_val = X_val / np.max(X_val, axis=1, keepdims=True) # Normalización igual que train

# 3. Cargar Test (96 Samples)
print("\n--- Cargando X_test (Samples Desconocidos) ---")
X_test_raw, _ = load_comp_data(os.path.join(path, '96_samples.csv'), is_train=False)
X_test = fix_val_test_shape(X_test_raw)
X_test = X_test / np.max(X_test, axis=1, keepdims=True) # Normalización igual que train

print(f"\nDimensiones finales:")
print(f"X_train: {X_train.shape}")
print(f"X_val:   {X_val.shape}")
print(f"X_test:  {X_test.shape}")
```

```python
# Concatenamos Train y Val para pasarlo al GridSearch
X_combined = np.vstack((X_train, X_val))
y_combined = np.vstack((y_train, y_val))

# Creamos el array de "test_fold"
# -1 indica: "Usar para entrenamiento"
# 0 indica: "Usar para validación"
train_indices = np.full((X_train.shape[0],), -1, dtype=int)
val_indices = np.full((X_val.shape[0],), 0, dtype=int)
test_fold = np.concatenate((train_indices, val_indices))

# PredefinedSplit fuerza al GridSearch a usar exactamente esta división
ps = PredefinedSplit(test_fold)

print(f"Estrategia de validación configurada.")
print(f"Indices de entrenamiento: {np.sum(test_fold == -1)}")
print(f"Indices de validación:    {np.sum(test_fold == 0)}")
```

```python
# Definimos los objetivos (targets)
targets = ['Glucose', 'Sodium Acetate', 'Magnesium Acetate']
predictions = {} # Diccionario para guardar predicciones

# --- MODELO 1: PLS Regression (Clásico y Robusto) ---
print("\n=== Entrenando PLS Regression ===")
pls_param_grid = {'n_components': [10, 20, 30, 40, 50]} # Grid de componentes

pls_preds_val = np.zeros(y_val.shape)
pls_preds_test = np.zeros((X_test.shape[0], 3))

for i, target in enumerate(targets):
    print(f"Optimizando PLS para {target}...")

    # Grid Search con PredefinedSplit
    grid = GridSearchCV(PLSRegression(), pls_param_grid, cv=ps, scoring='neg_mean_squared_error', n_jobs=-1)
    grid.fit(X_combined, y_combined[:, i])

    print(f"  Mejor n_components: {grid.best_params_}")

    # El mejor modelo ya está re-entrenado en X_combined (Train + Val) gracias a refit=True por defecto
    best_pls = grid.best_estimator_
    pls_preds_test[:, i] = best_pls.predict(X_test).flatten()
    pls_preds_val[:, i] = best_pls.predict(X_val).flatten() # Solo para ver error


# --- MODELO 2: XGBoost (Potencia de Gradient Boosting) ---
print("\n=== Entrenando XGBoost ===")
xgb_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.05, 0.1],
    'subsample': [0.8]
}

xgb_preds_val = np.zeros(y_val.shape)
xgb_preds_test = np.zeros((X_test.shape[0], 3))

for i, target in enumerate(targets):
    print(f"Optimizando XGBoost para {target}...")

    xgb = XGBRegressor(objective='reg:squarederror', n_jobs=-1, random_state=42)
    grid = GridSearchCV(xgb, xgb_param_grid, cv=ps, scoring='neg_mean_squared_error', n_jobs=-1)

    grid.fit(X_combined, y_combined[:, i])
    print(f"  Mejores params: {grid.best_params_}")

    best_xgb = grid.best_estimator_
    xgb_preds_test[:, i] = best_xgb.predict(X_test)
    xgb_preds_val[:, i] = best_xgb.predict(X_val)

print("\n¡Entrenamiento finalizado!")
```

```python
# Blending: 50% PLS + 50% XGBoost (Puedes ajustar estos pesos)
final_preds_test = (0.5 * pls_preds_test) + (0.5 * xgb_preds_test)
final_preds_val = (0.5 * pls_preds_val) + (0.5 * xgb_preds_val)

# Métricas de validación
print("\n=== Resultados en Validation (Transfer Plate) ===")
for i, target in enumerate(targets):
    mse = mean_squared_error(y_val[:, i], final_preds_val[:, i])
    r2 = r2_score(y_val[:, i], final_preds_val[:, i])
    print(f"{target}: MSE = {mse:.4f}, R2 = {r2:.4f}")

# Post-procesamiento: Eliminar negativos y Clipping inteligente
final_preds_test = np.maximum(final_preds_test, 0) # No puede haber concentraciones negativas

# Opcional: Recortar picos locos (Clipping basado en stats de train)
for i in range(3):
    max_val = np.percentile(y_train[:, i], 99.5) * 1.2 # Margen del 20% sobre el max visto
    final_preds_test[:, i] = np.clip(final_preds_test[:, i], 0, max_val)

print("\nPost-procesamiento completado.")
```

```python
# Crear DataFrame de envío
submission = pd.DataFrame()
submission['ID'] = range(1, 97) # IDs del 1 al 96
submission['Glucose'] = final_preds_test[:, 0]
submission['Sodium Acetate'] = final_preds_test[:, 1]
submission['Magnesium Sulfate'] = final_preds_test[:, 2] # Nota: A veces piden Magnesium Sulfate en vez de Acetate

# Verificación
print(submission.head())

# Guardar
submission.to_csv('submission_ensemble_transfer.csv', index=False)
print("\nArchivo 'submission_ensemble_transfer.csv' guardado con éxito.")
```

```python
def plot_metrics_heatmap(y_val, preds, target_names):
    # Calcular R2 y RMSE para cada target
    r2_scores = [r2_score(y_val[:, i], preds[:, i]) for i in range(len(target_names))]
    rmse_scores = [np.sqrt(mean_squared_error(y_val[:, i], preds[:, i])) for i in range(len(target_names))]

    # Crear DataFrame para visualización
    df_metrics = pd.DataFrame({'R2': r2_scores, 'RMSE': rmse_scores}, index=target_names)

    plt.figure(figsize=(6, len(target_names)*0.5 + 2))

    # Heatmap anotado
    sns.heatmap(df_metrics, annot=True, cmap="viridis", fmt=".3f", cbar=True)
    plt.title("Rendimiento del Modelo por Variable")
    plt.show()

# Llamada correcta a la función con tus datos
plot_metrics_heatmap(y_val, final_preds_val, targets)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# 1. Definir la función de reporte (Estilo Paper)
def reporte_calidad_cientifica(y_true, y_pred, nombre_variable, unidad="g/L"):
    """
    Calcula métricas y genera gráficas para un analito específico.
    Retorna un diccionario con las métricas para la tabla final.
    """
    # Cálculos
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    std_dev = np.std(y_true)
    rpd = std_dev / rmse if rmse > 0 else 0

    # Interpretación RPD
    if rpd > 3: cal = "Excelente"
    elif rpd > 2: cal = "Bueno"
    else: cal = "Pobre"

    # --- GRAFICAR ---
    sns.set_style("ticks")
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    # A. Predicho vs Medido
    ax[0].scatter(y_true, y_pred, edgecolor='k', alpha=0.6, s=40, color='#2b8cbe')
    min_val = min(y_true.min(), y_pred.min())
    max_val = max(y_true.max(), y_pred.max())
    ax[0].plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='1:1 Line')

    ax[0].set_xlabel(f'Referencia ({unidad})', fontsize=11)
    ax[0].set_ylabel(f'Predicho ({unidad})', fontsize=11)
    ax[0].set_title(f'{nombre_variable}: Predicho vs Referencia', fontsize=13, weight='bold')

    # Texto con estadísticas en el gráfico
    stats_text = f'$R^2$ = {r2:.3f}\nRMSE = {rmse:.3f}\nRPD = {rpd:.1f}'
    ax[0].text(0.05, 0.95, stats_text, transform=ax[0].transAxes,
               fontsize=10, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    ax[0].grid(True, linestyle='--', alpha=0.5)

    # B. Residuos
    residuos = y_true - y_pred
    ax[1].scatter(y_pred, residuos, edgecolor='k', alpha=0.6, s=40, color='#e6550d')
    ax[1].axhline(0, color='black', linestyle='--', lw=2)
    ax[1].set_xlabel(f'Valor Predicho ({unidad})', fontsize=11)
    ax[1].set_ylabel(f'Residuo ({unidad})', fontsize=11)
    ax[1].set_title(f'{nombre_variable}: Residuos', fontsize=13, weight='bold')
    ax[1].grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

    return {
        "Analito": nombre_variable,
        "R2": r2,
        "RMSE": rmse,
        "MAE": mae,
        "RPD": rpd,
        "Calidad": cal
    }

# 2. Ejecutar el bucle para las 3 variables
nombres_variables = ["Glucosa", "Acetato de Sodio", "Sulfato de Magnesio (MgSO4)"]
metricas_totales = []

print("GENERANDO REPORTES INDIVIDUALES...\n" + "="*40)

for i, nombre in enumerate(nombres_variables):
    # Extraemos los datos correspondientes a la columna i
    # Asegúrate que y_val y final_preds_val tienen la forma (n_muestras, 3)
    datos_reales = y_val[:, i]
    datos_predichos = final_preds_val[:, i]

    metrica = reporte_calidad_cientifica(datos_reales, datos_predichos, nombre)
    metricas_totales.append(metrica)

# 3. Crear Tabla Resumen para el Paper
print("\n" + "="*60)
print("TABLA RESUMEN PARA INCLUIR EN EL ARTÍCULO")
print("="*60)

df_resultados = pd.DataFrame(metricas_totales)
# Redondear para presentación limpia
df_resultados = df_resultados.round({'R2': 4, 'RMSE': 4, 'MAE': 4, 'RPD': 2})

display(df_resultados)

# Opcional: Exportar a Excel si lo necesitas
# df_resultados.to_excel("tabla_resultados_paper.xlsx", index=False)
```
